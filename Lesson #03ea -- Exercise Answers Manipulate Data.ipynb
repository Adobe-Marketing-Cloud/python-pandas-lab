## Advanced

Filter down the report to pageviews > 1000

```python
report.dataframe[report.dataframe.pageviews > 1000]
```

Plot a scatter chart of pageviews and visits to each page

```python
report.dataframe.plot(kind='scatter',x='pageviews',y='visits')
```

Create a new metric of pageivews per visit (which one has the highest?)

```python
report.dataframe['pv/visit'] = report.dataframe['pageviews']/report.dataframe['visits']
```

```python
report.dataframe.sort('pv/visit', ascending=False)
```
