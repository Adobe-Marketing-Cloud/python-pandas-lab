# Optional

Run a ranked report

```python
suite.report.element("page")\
    .metric("pageviews")\
    .metric("visits")\
    .metric("visitors")\
    .range("2015-02-01","2015-02-28")\
    .run()
```

Run a trended report

```python
suite.report.element("page")\
    .metric("pageviews")\
    .metric("visits")\
    .metric("visitors")\
    .granularity("day")\
    .range("2015-02-01","2015-02-28")\
    .run()
```

Breakdown by Geo State

```python
suite.report.element("page")\
    .breakdown("geostate")\
    .metric("pageviews")\
    .metric("visits")\
    .metric("visitors")\
    .granularity("day")\
    .range("2015-02-01","2015-02-28")\
    .run()
```
